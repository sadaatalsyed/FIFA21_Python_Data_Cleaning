### 11. Contract column

```python
fifa21['Contract'].dtype
```

```python
fifa21['Contract'].unique()
```

```python
for index,row in fifa21.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
# lets segrigate contrac column into start_date,end_date,contract length
def extract_contract_info(contract): 
    if (contract=='Free') or 'On Loan' in contract:
        start_date=np.nan
        end_date=np.nan
        contract_length=0
    else:
        start_date, end_date=contract.split(' ~ ')
        start_year= int(start_date[:4])
        end_year= int(end_date[:4])    
        contract_length=end_year - start_year
    return start_date, end_date, contract_length

# apply above function to contract column and create new columns
new_cols=['Contract Start','Contract End','Contract Length(years)'] # new columns to be inserted in dataset
new_data=fifa21['Contract'].apply(lambda x: pd.Series(extract_contract_info(x))) # converting data return from function into series
for i in range(len(new_cols)):
    fifa21.insert(loc=fifa21.columns.get_loc('Contract')+1+i,column=new_cols[i], value=new_data[i])
    
    
```

```python
fifa21.sample(4)
```

```python
def categorize_contract_status(contract):
    if contract=="Free":
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contact'

fifa21.insert(fifa21.columns.get_loc('Contract Length(years)')+1, 'Contract Status', fifa21['Contract'].apply(categorize_contract_status))
fifa21.sample(6)
```

### 13. Height

```python
fifa21['Height'].dtype
```

```python
fifa21['Height'].unique()
```

```python
# above output shows that some values are in cm and some in inches and feet so we need to convert all to same format so that column type can be set to integers
def convert_units(height):
    if 'cm' in height:
        return int(height.strip("cm"))
    else:
        feet, inches= height.split("'")
        total_inches=int(feet)*12+int(inches.strip(' " '))
        return round(total_inches*2.54)

fifa21['Height']=fifa21['Height'].apply(convert_units)
fifa21['Height'].unique()
```

```python
fifa21=fifa21.rename(columns={'Height': "Height(cm)"})
fifa21.sample(3)
```

### 14. Weight

```python
# unique method will give units in vlaues and data type  of the column
fifa21['Weight'].unique()
```

```python
# above output shows that some values are in kg and some in lbs so we need to convert all to same format so that column type can be set to integers
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip("kg"))
    else:
        
        pounds=int(weight.strip('lbs'))
        return round(pounds/2.205)
#apply function to weight column
fifa21['Weight']=fifa21['Weight'].apply(convert_weight)
fifa21['Weight'].unique()
```

```python
fifa21=fifa21.rename(columns={'Weight': "Weight(kg)"})
fifa21.sample(3)
```
